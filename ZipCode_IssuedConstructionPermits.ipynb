```python
import pandas as pd
import numpy as np
import matplotlib

import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
```

```python
# Store filepath in a variable
file = "Issued_Construction_Permits_New.csv"
```

```python
file_df = pd.read_csv(file)
file_df
```

```python
new_df = file_df[['FiscalYearIssued','OriginalZip']] 
new_df
```

```python
final_df = pd.DataFrame(new_df.groupby(['OriginalZip'])['OriginalZip'].count())
final_df.sort_values(by = ["OriginalZip"],ascending=False)
```

```python
# renamed_final_df = final_df.rename(columns={'OriginalZip':'OriginalZip_Count'})
# renamed_final_df
```

```python
zipcodes = final_df.index.values
zipcodes
```

```python
zipcodes = zipcodes.astype(int) % 100
zipcodes
```

```python
# zips = zipcodes['OriginalZip'].unique()
# zips

# final_zip = zips.astype(str)
# final_zip
```

```python
plt.figure(figsize=(60,20))

p1 = plt.bar(zipcodes, final_df["OriginalZip"])

plt.ylabel('Construction Permits')
plt.xlabel('Zipcodes')
plt.title('Construction Permits by Zipcode')
plt.xticks(zipcodes)

plt.show()
```
