```python
import pandas as pd
import numpy as np
import matplotlib

import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
```

```python
# Store filepath in a variable
file = "Issued_Construction_Permits_New.csv"
```

```python
file_df = pd.read_csv(file)
file_df
```

```python
new_df = file_df[['FiscalYearIssued','PermitClassMapped']]
new_df
```

```python
final_df = pd.DataFrame(new_df.groupby(['FiscalYearIssued','PermitClassMapped'])['PermitClassMapped'].count())
final_df
```

```python
Commercial_df = new_df[new_df["PermitClassMapped"]=="Commercial"]

final_commercial_df = pd.DataFrame(Commercial_df.groupby(['FiscalYearIssued'])['PermitClassMapped'].count())
final_commercial_df
```

```python
Residential_df = new_df[new_df["PermitClassMapped"]=="Residential"]

final_residential_df = pd.DataFrame(Residential_df.groupby(['FiscalYearIssued'])['PermitClassMapped'].count())
final_residential_df
```

```python
years = final_residential_df.index.values
years
```

```python
N = 7
commercial = (904,949,922,1062,1325,3211,5725)
residential = (1171,1895,1965,2192,2988,8124,14593)

ind = np.arange(len(commercial))
width = 0.35

p1 = plt.bar(ind, residential, width)
p2 = plt.bar(ind, commercial, width, bottom=residential)

plt.ylabel('Number of Permits')
plt.xlabel('Years')
plt.title('Austin Permits')
plt.xticks(ind, ('2012', '2013', '2014', '2015','2016','2017','2018'))
plt.legend((p1[0], p2[0]), ('Residential', 'Commercial'))

plt.show()
```
