```python
import matplotlib.pyplot as plt
import folium
import os
import pandas as pd
```

```python
geo_zip = "austin_area_zip_codes.geojson"
parks_path = os.path.join("../Resources/Clean/parks_by_zip.csv")
parks_df = pd.read_csv(parks_path)
parks_df["Zip Code"]= parks_df["Zip Code"].astype(str)
```

```python
austin_map = folium.Map(location=[30.277187,-97.736659],zoom_start=11)
# austin_map

austin_map.choropleth(geo_data=geo_zip,
                      data=parks_df,columns=["Zip Code", "Number of Parks"],
                      key_on= "feature.properties.zipcode",
                      legend_name= "Number of Parks",
                      fill_color="BuPu",
                      fill_opacity =0.8,
                      line_opacity=0.3)
austin_map
```
